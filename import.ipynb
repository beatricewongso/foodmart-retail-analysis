## duplicate values

#### full row duplicates

```python
# 1. transactions
print(f"Transactions - duplicates: {transactions.drop(columns=['transaction_id']).duplicated().sum()}")

# 2. customer
print(f"Customer - duplicates: {customer.drop(columns=['customer_id']).duplicated().sum()}")

# 3. product
print(f"Product - duplicates: {product.drop(columns=['product_id']).duplicated().sum()}")

# 4. store
print(f"Store - duplicates: {store.drop(columns=['store_id']).duplicated().sum()}")

# 5. region
print(f"Region - duplicates: {region.drop(columns=['region_id']).duplicated().sum()}")
```

```python
mask = transactions.drop(columns=['transaction_id']).duplicated(keep='first')
transactions = transactions[~mask].reset_index(drop=True)
```

```python
mask = transactions.drop(columns=['transaction_id']).duplicated(keep=False)
print(transactions[mask])
```

```python
# recreate transaction_id to account for the dropped duplicates 
transactions['transaction_id'] = transactions.index + 1
```

#### primary key duplicates

```python
# 1. transactions_id duplicates
print(f"Duplicate transaction_id: {transactions.duplicated(subset=['transaction_id']).sum()}")

# 2. customer_id duplicates
print(f"Duplicate customer_id: {customer.duplicated(subset=['customer_id']).sum()}")   

# 3. product_id duplicates
print(f"Duplicate product_id: {product.duplicated(subset=['product_id']).sum()}")

# 4. store_id duplicates
print(f"Duplicate store_id: {store.duplicated(subset=['store_id']).sum()}")

# 5. region_id duplicates
print(f"Duplicate region_id: {region.duplicated(subset=['region_id']).sum()}")
```

#### region duplicates

## NA Values

### checking for null values

```python
tables = ['store_performance', 'store', 'customer', 'store_products', 'regions']

null_summary = []
for table in tables:
    df = pd.read_sql(f"SELECT * FROM {table}", engine)
    nulls = df.isnull().sum()
    nulls = nulls[nulls > 0]  # only show columns with nulls
    for col, count in nulls.items():
        null_summary.append({'table': table, 'column': col, 'null_count': count})

pd.DataFrame(null_summary).sort_values(['table', 'null_count'], ascending=[True, False])
```
